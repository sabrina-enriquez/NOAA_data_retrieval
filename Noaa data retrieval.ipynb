```python
def get_month_url( station_id, product_name, begin_date='20210101', end_date='20210201' ):
    data = [('begin_date', begin_date),('end_date',end_date),('station',station_id),('product' , product_name),
        ('datum','mllw'),('units','metric'), ('time_zone','gmt'),('application','ports_screen'),('format','json')]
    url_values = urllib.parse.urlencode(data)
    url='https://api.tidesandcurrents.noaa.gov/api/prod/datagetter?'
    full_url=url+ url_values
    return full_url
    
```

```python
def get_data_water_level(full_url, state):
    response = urllib.request.urlopen(full_url)
    data = json.load(response)
    num_rows=len(data.get('data'))
    meta=data.get('metadata')
    new_col_names=list(meta.keys())
    new_cols={col_name: [meta.get(col_name)]*num_rows for col_name in new_col_names}
    df=pd.DataFrame(data.get('data'))
    for col_name in new_col_names:
        df[col_name]=new_cols.get(col_name)
    df['state']=[state]*num_rows
    df=df.rename(columns={'v':'v_water_level', 's':'s_water_level', 'f': 'f_water_level', 'q':'q_water_level'})
    return df
```

```python
def get_data_prediction(full_url, station):
    response = urllib.request.urlopen(full_url)
    data = json.load(response)
    num_rows=len(data.get('predictions'))
    df=pd.DataFrame(data.get('predictions'))
    df['id']=[station]*num_rows
    df=df.rename(columns={'v':'v_prediction'})
    return df
```

```python
#function to take in state and station dict, and save path
def save_state_merged_df(state_station_dict, state, start_date, finish_date, save_path="/Volumes/Seagate Backup Plus Drive/NOAA data/" ):
    for station in state_station_dict.get(state):
        first_df=get_data_water_level(full_url=get_month_url(station_id=station, product_name=products[0], begin_date= start_date, end_date=finish_date), state=state)
        second_df=get_data_prediction(full_url=get_month_url(station_id=station, product_name=products[2], begin_date=start_date, end_date=finish_date), station=station)
        joined=first_df.merge(second_df, on=['t','id'])
        joined.to_csv(f"{save_path}{state}/{start_date}/{station}_{start_date}.csv", index=False)
        print(f"saved to: {save_path}{state}/{start_date}/{station}_{start_date}.csv")
    return 0
    
```

```python
def dates_list_builder(first_month, last_month):
    dates=[f"{first_month}"]
    next_year=first_month+10000
    while(first_month<last_month):
        first_month=first_month+100
        dates.append(f"{first_month}")
    dates.append(f"{next_year}")
    return dates
        
    
```
